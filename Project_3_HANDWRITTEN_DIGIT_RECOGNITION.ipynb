```python
# ============================================================
# HANDWRITTEN DIGIT RECOGNITION
# Dataset: MNIST (built into TensorFlow/Keras - no download needed)
# Model: CNN (LeNet-5 inspired architecture)
# ============================================================

# IMPORTING LIBRARIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow import keras
from tensorflow.keras import datasets
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout, AveragePooling2D
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, ConfusionMatrixDisplay

# ============================================================
# STEP 1 — LOAD DATASET
# MNIST is built into Keras — no file needed
# 60000 training images, 10000 test images
# Each image is 28x28 pixels, grayscale
# ============================================================
print("="*50)
print("STEP 1 — LOADING DATASET")
print("="*50)

(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()

print("Dataset loaded successfully!")
print("Training set shape :", x_train.shape)
print("Testing set shape  :", x_test.shape)
print("Labels             :", np.unique(y_train))

# ============================================================
# STEP 2 — EDA
# Visualize sample digits and class distribution
# ============================================================
print("\n" + "="*50)
print("STEP 2 — EXPLORATORY DATA ANALYSIS")
print("="*50)

# Show sample digits
plt.figure(figsize=(10, 4))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(f"Label: {y_train[i]}")
    plt.axis('off')
plt.suptitle('Sample Images from MNIST Dataset')
plt.tight_layout()
plt.show()

# Class distribution
plt.figure(figsize=(7, 4))
sns.countplot(x=y_train, hue=y_train, palette='Set2', legend=False)
plt.title('Class Distribution — Digit Counts in Training Set')
plt.xlabel('Digit')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

print("\nClass Distribution:")
unique, counts = np.unique(y_train, return_counts=True)
for digit, count in zip(unique, counts):
    print(f"  Digit {digit} : {count} samples")

# ============================================================
# STEP 3 — PREPROCESSING
# Reshape, normalize and split into train/val/test
# ============================================================
print("\n" + "="*50)
print("STEP 3 — PREPROCESSING")
print("="*50)

# Reshape for CNN input (add channel dimension)
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test  = x_test.reshape(x_test.shape[0], 28, 28, 1)

# Validation split
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.1, random_state=42)

# Normalize using mean and std
x_train = (x_train - x_train.mean()) / x_train.std()
x_val   = (x_val   - x_val.mean())   / x_val.std()
x_test  = (x_test  - x_test.mean())  / x_test.std()

# One-hot encode labels
num_labels = 10
y_train_cat = to_categorical(y_train, num_classes=num_labels)
y_val_cat   = to_categorical(y_val,   num_classes=num_labels)
y_test_cat  = to_categorical(y_test,  num_classes=num_labels)

print("Preprocessing done!")
print("Train shape      :", x_train.shape)
print("Validation shape :", x_val.shape)
print("Test shape       :", x_test.shape)

# ============================================================
# STEP 4 — BUILD MODEL
# LeNet-5 inspired CNN architecture
# Conv2D → AveragePooling → Conv2D → AveragePooling → Dense
# ============================================================
print("\n" + "="*50)
print("STEP 4 — BUILDING MODEL")
print("="*50)

model = Sequential([
    Conv2D(6,  kernel_size=(5,5), strides=(1,1), activation='tanh', input_shape=(28,28,1)),
    AveragePooling2D(pool_size=(2,2), strides=(1,1)),
    Conv2D(16, kernel_size=(5,5), strides=(1,1), activation='tanh'),
    AveragePooling2D(pool_size=(2,2), strides=(2,2)),
    Flatten(),
    Dense(84, activation='tanh'),
    Dense(num_labels, activation='softmax')
])

model.summary()

# ============================================================
# STEP 5 — COMPILE AND TRAIN
# Loss: categorical crossentropy (multi-class classification)
# Optimizer: Adam
# Epochs: 15 (increased from 5 for better convergence)
# ============================================================
print("\n" + "="*50)
print("STEP 5 — TRAINING MODEL")
print("="*50)

model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

history = model.fit(
    x_train, y_train_cat,
    batch_size=32,
    epochs=15,
    validation_data=(x_val, y_val_cat)
)

# ============================================================
# STEP 6 — TRAINING CURVES
# Plot accuracy and loss for train vs validation
# Helps identify overfitting or underfitting
# ============================================================
print("\n" + "="*50)
print("STEP 6 — TRAINING CURVES")
print("="*50)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Accuracy
axes[0].plot(history.history['accuracy'],     label='Train Accuracy')
axes[0].plot(history.history['val_accuracy'], label='Val Accuracy')
axes[0].set_title('Model Accuracy')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend()

# Loss
axes[1].plot(history.history['loss'],     label='Train Loss')
axes[1].plot(history.history['val_loss'], label='Val Loss')
axes[1].set_title('Model Loss')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend()

plt.tight_layout()
plt.show()

# ============================================================
# STEP 7 — EVALUATION
# Test accuracy, Classification Report, Confusion Matrix
# ============================================================
print("\n" + "="*50)
print("STEP 7 — EVALUATION")
print("="*50)

score = model.evaluate(x_test, y_test_cat, verbose=0)
print(f"Test Loss     : {score[0]:.4f}")
print(f"Test Accuracy : {score[1]:.4f}")

# Classification report
y_pred = np.argmax(model.predict(x_test), axis=1)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Confusion matrix
plt.figure(figsize=(10, 8))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
plt.title('Confusion Matrix — CNN')
plt.tight_layout()
plt.show()

# ============================================================
# STEP 8 — SAMPLE PREDICTIONS
# Show actual vs predicted labels on test images
# ============================================================
print("\n" + "="*50)
print("STEP 8 — SAMPLE PREDICTIONS")
print("="*50)

plt.figure(figsize=(12, 5))
for i in range(10):
    plt.subplot(2, 5, i+1)
    # Reshape back to 28x28 for display
    img = x_test[i].reshape(28, 28)
    plt.imshow(img, cmap='gray')
    actual    = y_test[i]
    predicted = y_pred[i]
    color = 'green' if actual == predicted else 'red'
    plt.title(f"A:{actual} P:{predicted}", color=color)
    plt.axis('off')
plt.suptitle('Sample Predictions — Green=Correct, Red=Wrong')
plt.tight_layout()
plt.show()

# ============================================================
# FINAL SUMMARY
# ============================================================
print("\n" + "="*50)
print("FINAL SUMMARY")
print("="*50)
print(f"Model         : CNN (LeNet-5 inspired)")
print(f"Dataset       : MNIST (70,000 images)")
print(f"Test Accuracy : {score[1]*100:.2f}%")
print(f"Epochs        : 15")
print(f"Batch Size    : 32")
```
